```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.sample()
```

```python
df['target']=iris.target
df.sample()
```

```python
df['target_names']= df['target'].apply(lambda x: iris.target_names[x])
```

```python
df.sample()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flower_count = df.target_names.value_counts()
flower_count.plot(kind='bar',color='skyblue',edgecolor='black')
plt.xlabel('Flower Names',labelpad=10)
plt.ylabel('Flower Count',labelpad=10)
plt.xticks(rotation=360)
plt.show()
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
```

```python
df0.sample()
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='purple',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='skyblue',marker='*')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='purple',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='skyblue',marker='*')
plt.xlabel('petal length')
plt.ylabel('petal width')
```

```python
X= df.drop(['target','target_names'], axis='columns')
X
```

```python
y=df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.svm import SVC
model=SVC(C=3.0)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
newmodel=SVC(C=3,gamma='scale',kernel='rbf')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Xdig,ydig, test_size=0.2)
```

```python
newmodel.fit(X_train,y_train)
```

```python
newmodel.score(X_test,y_test)
```
