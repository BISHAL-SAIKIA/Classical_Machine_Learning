```python
df = df.astype(float)
```

```python
df.dropna(inplace=True)
```

```python
X=df.drop(['Price'],axis=1)
y=df.Price
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg  = LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

## Using Lasso (L1 Regularized) Regression Model

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

## Using Ridge (L2 Regularized) Regression Model

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=50, max_iter=100,tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```

## Using L1 AND L2 Regularization increases the penalties on errors and reduces overfitting by introducing a new term in the equation. Hence it produces significantly same score for training as well as testing data.
