```python
import pandas as pd
df=pd.read_csv('spam.csv')
```

```python
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam']=df.Category.apply(lambda x:1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message,df.spam,test_size=0.2)
```

### Now since the messages are in textual form, we need to convert it into numerical values so that it can used for model training. And to do that, we use sklearn countVectorizer.Countvectorizer counts the total unique words and converts a sentence into vector with word count for different word count corresponding to the given word

![Alt text](countVector.png)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v=CountVectorizer()
X_train_count= v.fit_transform(X_train.values)
```

```python
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

# Sklearn Pipeline

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
